```python
class cubicfurniture:
    def __init__(self, d, w, h, color="blanc", front="w",price=0):
        self.d = d
        self.w = w
        self.h = h
        self.front = front
        self.price = price
        self.color = color
        
    def properties(self):
        print( "depth = ", self.d)
        print( "width= ", self.w)
        print( "length = ", self.h)
        print( "front = ", self.front)
        print( "price = ", self.price)
        print( "color = ", self.color)
    
    
```
